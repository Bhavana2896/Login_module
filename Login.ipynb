```python
import re
import pickle
import os
class User:
  #dictionary defined outside function to be accessible to all functions
  user_data = {}
  
  # else:
  #   print("File Exists")    
 
  # To read multiple values from a pickled file until EOF error is raised
  # Unpickling
  def loadall(self, filename):
    with open(filename, "rb") as f:
        while True:
            try:
                yield pickle.load(f)
            except EOFError:
                break
  
  #Creating a dictionary for username & password
  def load_data(self):
    items = self.loadall("user_db1.txt")
    for data in items:
      a,b=data.split(", ")
      b=b.strip()#removes extra spaces
      self.user_data[a] = b
  
  #User Registration 
  def register(self):
    #create file if it doesn't exist
    fp = open(r'/content/user_db1.txt','ab')
    fp.close()
    #calling function load_data to check if user already exists
    self.load_data() 
    username=input("Create Username: ")
    password=input("Create password: ")
    confirm_password=input("Re-enter password: ")

    if username in self.user_data.keys():
      print("Username already exists!!")
      self.register()
    elif password!=confirm_password:
      print("Passwords don't match")
      self.register()
    elif len(password)<8:
      print("Password is too short")
      self.register()
    elif re.search('[\s]',password):
      print("Please remove any spaces included in password")
      self.register()
    elif not re.search('[a-zA-Z]',password):
      print("Password length must be atleast 8 with alphanumeric & special characters(!@#$%^&*).Space must not be included!!")
      self.register()
    elif not re.search('[0-9]',password):
      print("Password length must be atleast 8 with alphanumeric & special characters(!@#$%^&*).Space must not be included!!")
      self.register()
    elif not re.search('[!@#$%^&*]',password):
      print("Password length must be atleast 8 with alphanumeric & special characters(!@#$%^&*).Space must not be included!!")
      self.register()
    else:
      #Pickling
      #appending data to pickle file in username, password format. File created if not available
      fw=open("user_db1.txt","ab")
      pickle.dump((username+", "+password+"\n"), fw)
      fw.close()
      print("User registration successful!!!")

    #User login option after successful registration
    while True:
      user_input = input('Do you want to Log in? yes/no: ')

      if user_input.lower() == 'yes':
          self.access()
          break
      elif user_input.lower() == 'no':
          break


  def access(self):
    self.load_data();
    username=input("Enter Username: ")
    password=input("Enter password: ")

    #if username non-existent, error shoots up so try is reqd
    try:
      if self.user_data[username]:
        if password==self.user_data[username]:
          print("Login Successful")
          print("Hi",username)
        else:
          print("Incorrect username or password")
          self.access()
      else:
        print("Username or password doesn't exist")
        self.access()
    except:
      print("Username or password doesn't exist")
      self.access()
    
#Home page for users to choose either login or signup
def home(option=None):
  user = User()
  option=input("Login | Signup: ")
  if option=="Login":
    user.access()
  elif option=="Signup":
    user.register()
  else:
    print("Please enter an option")
    home()

home()
```
